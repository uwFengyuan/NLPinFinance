```python
import json
import pickle
import gzip
import os
```

```python
cloth_filename = '/data/liufengyuan/NLPinFinance/Unziped_Filtered_Data/Clothing_Shoes_and_Jewelry.pkl'
```

```python
cloth_data = pickle.load(open(cloth_filename,'rb'))
```

```python
print(cloth_data[0])
```

```python
cloth_ids = []
for data in cloth_data:
    cloth_ids.append(data['asin'])
print(len(cloth_ids))
```

```python
print(cloth_ids[:3])
```

```python
karpathy_split_train_path = '/data/liufengyuan/NLPinFinance/LFYdata/annotations/karpathy_split_train.json'
karpathy_split_val_path = '/data/liufengyuan/NLPinFinance/LFYdata/annotations/karpathy_split_val.json'
```

```python
karpathy_split_train = json.load(open(karpathy_split_train_path,'r'))
karpathy_split_val = json.load(open(karpathy_split_val_path,'r'))
```

```python
karpathy_split_val
```

```python
print(len(karpathy_split_val['images']))
```

```python
cloth_karpathy_split_val = {'type': 'caption',
                            'images':[]}
```

```python
for idx,val_image in enumerate(karpathy_split_val['images']):
    val_image_id = val_image['id']
    if val_image_id in cloth_ids:
        cloth_karpathy_split_val['images'].append(val_image)
    if idx % 20 == 0:
        print(idx)
```

```python
print(len(cloth_karpathy_split_val['images']))
```

```python
json.dump(cloth_karpathy_split_val,open('/data/liufengyuan/NLPinFinance/LFYdata/annotations/cloth_karpathy_split_val.json','w'))
```
